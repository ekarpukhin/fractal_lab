```python
import holoviews as hv
from logic import fractal
hv.extension('bokeh')
```

```python
seq = (0,1,1,1,2,0) # sequence, which fractal is build on (coded ABBBCA). Thus, mine fractal has 3 dimentions
h, w = 600, 600 # height and width of picture
N = 100 # number of iterations, time of refreshing around 1 sec
xmin, xmax = 0, 4 # x range of picture
ymin, ymax = 0, 4 # y range of picture
```

```python
def graph(x_range, y_range):
    '''
    Given coor-s of part of the picture that we've choosen, returns this part of picture in better quality
    '''
    z = 0.01 # there we choosing z-coord, because we can't see full 3-dim picture 
    x0, x1 = x_range
    y1, y0 = y_range # inverting y_range, because array filling from top and left, but we want it starting from bottom and left to correctly draw picture
    return hv.Image(fractal(x0, x1, y0, y1, z, h, w, seq, N), bounds=(x0, y0, x1, y1))
```

```python
# creating dynamic map
range_xy = hv.streams.RangeXY(x_range=(xmin,xmax), y_range=(ymin,ymax))
dmap = hv.DynamicMap(graph, streams=[range_xy])
dmap.opts(hv.opts.Image(height=h, width=w, cmap='twilight', colorbar=True))
```
